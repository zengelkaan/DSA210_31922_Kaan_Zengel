## INTRODUCTION

This project aims to explore the factors influencing football team performance in the Premier League 2023-2024 season. By analyzing official team statistics such as squad value, age, possession, and defensive metrics, we attempt to determine which variables significantly impact the total number of points a team earns. Using data science tools, statistical hypothesis testing, and machine learning models, the study provides insights into how different metrics affect outcomes and whether predictive models can estimate a team's success.

---

## WHAT DID I DO?

Using a dataset containing statistics from all Premier League teams in the 2023-2024 season, I:

- Preprocessed the data by engineering new features such as "Conceded Per Match" and "Log Squad Value"
- Created visualizations to explore distributions and correlations
- Performed Pearson correlation tests to identify statistically significant relationships with team performance (measured by Points)
- Trained and evaluated 4 machine learning models: Linear Regression, Decision Tree, Random Forest, and XGBoost
- Selected the best model based on R² score
- Visualized actual vs predicted values for performance validation

---

## KEY VARIABLES ANALYZED

- **Log Squad Value:** Log-transformed market value of the team
- **Foreign Player Ratio:** Proportion of non-local players in the team
- **Conceded Per Match:** Goals conceded per match
- **Goals Per Match:** Goals scored per match
- **Goal Difference:** Difference between goals scored and conceded
- **Possession Efficiency:** A composite metric of ball possession and effectiveness
- **Squad Efficiency:** Performance relative to financial value
- **Average Age:** Mean age of the players

---

## HYPOTHESIS TESTING INSIGHTS

### 1. Log Squad Value vs Points
- **Correlation Coefficient:** 0.90
- **P-value:** 0.0000 
- **Interpretation:** Strong, statistically significant positive relationship. 
- → *"When a team's Log Squad Value increases, their total points tend to increase as well."*

### 2. Foreign Player Ratio vs Points
- **Correlation Coefficient:** -0.19
- **P-value:** 0.4178
- **Interpretation:** No statistically significant relationship.
- → *"Having more or fewer foreign players does not appear to affect total points."*

### 3. Conceded Per Match vs Points
- **Correlation Coefficient:** -0.83
- **P-value:** 0.0000
- **Interpretation:** Strong, statistically significant negative relationship.
- → *"When teams concede more goals per match, their total points tend to decrease."*

### 4. Possession Efficiency vs Points
- **Correlation Coefficient:** -0.11
- **P-value:** 0.6559
- **Interpretation:** No significant relationship.
- → *"Possession alone does not guarantee better team performance."*

## MACHINE LEARNING RESULTS

Trained four models on 80% of the dataset and tested on 20%:

| Decision Tree      |  48.75              | 0.93    |
| Random Forest      | 144.25              | 0.80       |
| XGBoost            | 53.73               | 0.93      |

> **Best Model:** *Decision Tree* with an R² score of *[0.93]*

The model was able to effectively predict team points based on selected features. Actual vs predicted scatter plots were generated to validate model fit.

## INSIGHTS AND UNDERSTANDING

- **Strong predictors** of team performance include squad value and defensive strength (low goals conceded per match).
- **Possession metrics** alone are not sufficient to explain performance.
- **Foreign player ratio** and **average age** were not significant contributors to team success.
